#### SCATTER PLOT:- TO FIND THE RELATIONSHIP BETWEEN TWO VARAIBALES

```python
def relational_graph(df):
    """ relational graph(scatter plot) between pental_length and sepal_length"""
    plt.figure(figsize=(6, 4))
sns.scatterplot(x='petal_length', y='sepal_length', hue='species', data=df)
plt.xlabel('Petal_Length')
plt.ylabel('sepal_length')
plt.title('Petal Length vs. sepal length by Species')
plt.show()
```

```python
def categorical_graph(df):
    """categorical graph(box plot) between sepal width and petal width"""
    plt.figure(figsize(4,6))
sns.boxplot(x='species',y='petal_width',data=df)
plt.title("distribution of petal width")
plt.xlabel('species')
plt.ylabel('petal_width')
plt.show()
```

```python
def stastical_graph(df):
    """stastical graph(histrogram) of sepal width of species"""
    plt.figure(figsize=(4,6))
sns.histplot(x='sepal_width',hue='species',data=df)
plt.title('distribution of sepal width')
plt.xlabel('sepal_width')
plt.ylabel('frequency')
plt.show()
```

```python
def des_data(df):
    """Describes the dataset,which includs  std, min,mean, max, median, skewness, and kurtosis for numerical columns."""
    num_df = df.select_dtypes(include='number')
    
    # descriptive stats
    description = num_df.describe().T  
    description['median'] = num_df.median()  
    description['skewness'] = num_df.apply(skew)  
    description['kurtosis'] = num_df.apply(kurtosis)  
    
    return description
def correlation_matrix(df):
    """the correlation matrix for numeric columns."""
    num_df = df.select_dtypes(include='number')
    correlation_matrix = num_df.corr()
    plt.figure(figsize=(6, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
    plt.title("Correlation Matrix of Iris Dataset")
    plt.show()


iris = sns.load_dataset('iris') 
description = des_data(iris)
print(description)

#correlation matrix
correlation_matrix(iris)
```
